```python
from pyspark.sql import functions as F

df_fixed = df.withColumn("price", F.col("price").cast("double"))
```

```python

df_fixed.select("brand", "price") \
    .groupBy("brand") \
    .agg(F.sum("price").alias("total_price")) \
    .orderBy(F.col("total_price").desc()) \
    .show()
```

## Average Price per Category

```python
df.groupBy("category_code") .agg(F.avg("price").alias("avg_price")).orderBy(F.col("avg_price").desc()).show() 
```

Remove NULL Categories Before Aggregatio

```python
df.filter(F.col("category_code").isNotNull()) \
  .groupBy("category_code") \
  .agg(F.sum("price").alias("total_sales")) \
  .orderBy(F.col("total_sales").desc()) \
  .show()
```
