```python
# Create simple DataFrame
data = [("iPhone", 999), ("Samsung", 799), ("MacBook", 1299)]
df = spark.createDataFrame(data, ["product", "price"])
df.show()

# Filter expensive products
df.filter(df.price > 1000).show()
```

# Adding new Columns

```python
from pyspark.sql.functions import *

df2 = df.withColumn(
    "category",
    when(df.price >= 1000, "Premium").otherwise("Standard")
)

df2.show()
```

```python


df3 = df2.withColumn(
    "discounted_price",
    when(col("category") == "Premium", col("price") * 0.9)
    .otherwise(col("price"))
)

df3.show()
```

# GroupBy

```python


df_grouped = df3.groupBy("category").agg(
    count("*").alias("product_count"),
    avg("price").alias("avg_price"),
    max("price").alias("max_price"),
    min("price").alias("min_price")
)

df_grouped.show()
```

```python
df_grouped.orderBy("avg_price", ascending=False).show()
```

```python
df_grouped.filter(col("avg_price") > 900).show()
```
