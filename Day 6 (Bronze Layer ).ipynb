###  Bronze Layer

```python
raw = spark.table('workspace.ecommerce.ecommerce_data')
```

```python
raw.columns
```

```python
import pyspark.sql.functions as F
```

```python
spark.sql("""
CREATE SCHEMA IF NOT EXISTS workspace.Bronze
""")
```

```python
spark.sql("""
CREATE VOLUME IF NOT EXISTS workspace.Bronze.ecommerce_data
""")
```

```python
raw.withColumn("ingestion_ts", F.current_timestamp()) \
   .write.format("delta") \
   .mode("overwrite") \
   .saveAsTable("workspace.Bronze.ecommerce_data")
```
