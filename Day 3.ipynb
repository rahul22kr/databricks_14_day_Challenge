```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window
```

###Rank products by price within each category

```python
window_spec = Window.partitionBy(
    "category_code"
).orderBy(
    col("price").desc()
)

```

### Rank

```python
df_rownum = df_oct.withColumn(
    "row_number",
    row_number().over(window_spec)
)

display(df_rownum)
```

```python
df_rank = df_oct.withColumn(
    "rank",
    rank().over(window_spec)
)

df_rank.display()
```

```python
df_dense = df_oct.withColumn(
    "dense_rank",
    dense_rank().over(window_spec)
)

df_dense.display()
```

```python
df_top3 = df_oct.withColumn(
    "dense_rank",
    dense_rank().over(window_spec)
).filter("dense_rank <= 3")
display(df_top3)
```

### Running Sum

```python
df_running = df_oct.withColumn(
    "running_price_total",
    sum("price").over(window_spec)
)

display(df_running)
```
